Task - 1:
Create a Python program that converts between different units of measurement.

• The program should:
1. Prompt the user to choose the type of conversion (e.g., length, weight, volume).
2. Ask the user to input the value to be converted.
3. Perform the conversion and display the result.
4. Handle potential errors, such as invalid input or unsupported conversion types.

• Requirements:
1. Functions: Define at least one function to perform the conversion.
2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values).
3. User Input: Prompt the user to select the conversion type and input the value.
4. Docstrings: Include a docstring in your function to describe its purpose, parameters, and
return value.

• Conversion Options:

1. Length:

– Convert meters (m) to feet (ft).

– Convert feet (ft) to meters (m).
2. Weight:

– Convert kilograms (kg) to pounds (lbs).

– Convert pounds (lbs) to kilograms (kg).

3. Volume:

– Convert liters (L) to gallons (gal).

– Convert gallons (gal) to liters (L).

```python
def convert(choice, value):
    """
    Convert units based on the user's choice.

    Parameters:
    choice (int): Conversion option selected by the user
    value (float): Value to convert

    Returns:
    float: Converted value
    """

    if choice == 1:
        return value * 3.28      # meters to feet
    elif choice == 2:
        return value / 3.28      # feet to meters
    elif choice == 3:
        return value * 2.2       # kg to pounds
    elif choice == 4:
        return value / 2.2       # pounds to kg
    elif choice == 5:
        return value * 0.26      # liters to gallons
    elif choice == 6:
        return value / 0.26      # gallons to liters


print("Unit Converter")
print("1. Meters to Feet")
print("2. Feet to Meters")
print("3. Kilograms to Pounds")
print("4. Pounds to Kilograms")
print("5. Liters to Gallons")
print("6. Gallons to Liters")

try:
    choice = int(input("Choose conversion (1-6): "))
    value = float(input("Enter value: "))

    result = convert(choice, value)
    print("Converted value:", result)

except:
    print("Invalid input. Please enter numbers only.")
```

Task - 2:
Create a Python program that performs various mathematical operations on a list of numbers.

• The Program should:
1. Prompt the user to choose an operation (e.g., find the sum, average, maximum, or minimum
of the numbers).
2. Ask the user to input a list of numbers (separated by spaces).
3. Perform the selected operation and display the result.
4. Handle potential errors, such as invalid input or empty lists.

• Requirements:
1. Functions: Define at least one function for each operation (sum, average, maximum, mini-
mum).
2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values or
empty lists).
3. User Input: Prompt the user to select the operation and input the list of numbers.
4. Docstrings: Include a docstring in each function to describe its purpose, parameters, and
return value.


```python
def find_sum(numbers):
    """
    Returns the sum of a list of numbers.

    Parameters:
    numbers (list): list of numeric values

    Returns:
    float: sum of the numbers
    """
    return sum(numbers)


def find_average(numbers):
    """
    Returns the average of a list of numbers.

    Parameters:
    numbers (list): list of numeric values

    Returns:
    float: average value
    """
    return sum(numbers) / len(numbers)


def find_max(numbers):
    """
    Returns the maximum number in a list.

    Parameters:
    numbers (list): list of numeric values

    Returns:
    float: maximum value
    """
    return max(numbers)


def find_min(numbers):
    """
    Returns the minimum number in a list.

    Parameters:
    numbers (list): list of numeric values

    Returns:
    float: minimum value
    """
    return min(numbers)


print("Math Operations Program")
print("1. Sum")
print("2. Average")
print("3. Maximum")
print("4. Minimum")

try:
    choice = int(input("Choose operation (1-4): "))

    user_input = input("Enter numbers separated by spaces: ")

    numbers = [float(x) for x in user_input.split()]

    if len(numbers) == 0:
        print("Error: The list cannot be empty.")

    elif choice == 1:
        print("Sum =", find_sum(numbers))

    elif choice == 2:
        print("Average =", find_average(numbers))

    elif choice == 3:
        print("Maximum =", find_max(numbers))

    elif choice == 4:
        print("Minimum =", find_min(numbers))

    else:
        print("Invalid choice.")

except ValueError:
    print("Error: Please enter valid numbers.")
```

1. Flatten a Nested List:
Write a Python function that takes a nested list and flattens it into a single list, where all the elements
are in a single dimension.

• Requirements:

– Define a function flatten(lst) that takes a nested list lst and returns a flattened version
of the list.

– Example: For the input [[1, 2], [3, 4], [5]], the output should be [1, 2, 3, 4, 5].

```python
def flatten(lst):
    """
    Flattens a nested list into a single list.

    Parameters:
    lst (list): A nested list (list of lists)

    Returns:
    list: A flattened list with all elements in a single dimension
    """
    flat_list = []
    for sublist in lst:
        for item in sublist:
            flat_list.append(item)
    return flat_list

nested_list = [[10, 20], [30, 40, 50], [60]]

result = flatten(nested_list)

print("Original nested list:", nested_list)
print("Flattened list:", result)
```

3. Sum of All Elements in a Nested List:

Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth).

• Requirements:

– Define a function sum nested(lst) that takes a nested list lst and returns the sum of all
the elements.

– Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 21.

```python
def sum_nested(lst):
    """
    Calculates the sum of all numbers in a nested list, regardless of depth.

    Parameters:
    lst (list): A nested list containing numbers or other nested lists

    Returns:
    int/float: Sum of all elements
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)  # recursive call for sublist
        else:
            total += item
    return total


nested_list = [[10, 20], [30, [40, 50]], 60]

result = sum_nested(nested_list)

print("Nested list:", nested_list)
print("Sum of all elements:", result)
```

4. Remove Specific Element from a Nested List:

Write a Python function that removes all occurrences of a specific element from a nested list.

• Requirements:

– Define a function remove element(lst, elem) that removes elem from lst and returns the
modified list.

– Example: For the input lst = [[1, 2], [3, 2], [4, 5]] and elem = 2, the output should
be [[1], [3], [4, 5]].

```python
def remove_element(lst, elem):
    """
    Removes all occurrences of a specific element from a nested list.

    Parameters:
    lst (list): A nested list
    elem (any): The element to remove

    Returns:
    list: Modified nested list with the element removed
    """
    new_list = []
    for sublist in lst:
        # Remove the element from each sublist
        filtered = [x for x in sublist if x != elem]
        new_list.append(filtered)
    return new_list

nested_list = [[10, 20], [30, 20], [40, 50]]
element_to_remove = 20

result = remove_element(nested_list, element_to_remove)

print("Original nested list:", nested_list)
print("Nested list after removing", element_to_remove, ":", result)
```

5. Find the Maximum Element in a Nested List:

Write a Python function that finds the maximum element in a nested list (regardless of depth).

• Requirements:

– Define a function find max(lst) that takes a nested list lst and returns the maximum
element.

– Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 6.

```python
def find_max(lst):
    """
    Finds the maximum element in a nested list, regardless of depth.

    Parameters:
    lst (list): A nested list containing numbers or other nested lists

    Returns:
    int/float: Maximum element in the nested list
    """
    max_value = float('-inf')  # Start with negative infinity
    for item in lst:
        if isinstance(item, list):
            max_value = max(max_value, find_max(item))  # Recursive call
        else:
            max_value = max(max_value, item)
    return max_value

nested_list = [[10, 20], [30, [40, 50]], 60]

result = find_max(nested_list)

print("Nested list:", nested_list)
print("Maximum element:", result)
```

6. Count Occurrences of an Element in a Nested List:

Write a Python function that counts how many times a specific element appears in a nested list.

• Requirements:

– Define a function count occurrences(lst, elem) that counts the occurrences of elem in
the nested list lst.

– Example: For the input lst = [[1, 2], [2, 3], [2, 4]] and elem = 2, the output should
be 3.

```python
def count_occurrences(lst, elem):
    """
    Counts the number of times a specific element appears in a nested list.

    Parameters:
    lst (list): A nested list
    elem (any): The element to count

    Returns:
    int: Number of occurrences of the element
    """
    count = 0
    for sublist in lst:
        for item in sublist:
            if item == elem:
                count += 1
    return count

nested_list = [[10, 20], [20, 30], [40, 20]]
element_to_count = 20

result = count_occurrences(nested_list, element_to_count)

print("Nested list:", nested_list)
print(f"Number of occurrences of {element_to_count}:", result)
```

7. Flatten a List of Lists of Lists:

Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.

• Requirements:

– Define a function deep flatten(lst) that takes a deeply nested list lst and returns a single
flattened list.

– Example: For the input [[[1, 2], [3, 4]], [[5, 6], [7, 8]]], the output should be
[1, 2, 3, 4, 5, 6, 7, 8].

```python
def deep_flatten(lst):
    """
    Flattens a deeply nested list into a single list, regardless of depth.

    Parameters:
    lst (list): A deeply nested list

    Returns:
    list: A flattened list containing all elements
    """
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(deep_flatten(item))  # recursive call
        else:
            flat_list.append(item)
    return flat_list

nested_list = [[[10, 20], [30, 40]], [[50, 60], [70, 80]]]

result = deep_flatten(nested_list)

print("Original nested list:", nested_list)
print("Deep flattened list:", result)
```

8. Nested List Average:

Write a Python function that calculates the average of all elements in a nested list.

• Requirements:

– Define a function average nested(lst) that takes a nested list lst and returns the average
of all the elements.

– Example: For the input [[1, 2], [3, 4], [5, 6]], the output should be 3.5.

```python
def average_nested(lst):
    """
    Calculates the average of all elements in a nested list, regardless of depth.

    Parameters:
    lst (list): A nested list containing numbers

    Returns:
    float: Average of all elements
    """
    def flatten_and_sum(lst):
        total = 0
        count = 0
        for item in lst:
            if isinstance(item, list):
                sub_total, sub_count = flatten_and_sum(item)
                total += sub_total
                count += sub_count
            else:
                total += item
                count += 1
        return total, count

    total_sum, total_count = flatten_and_sum(lst)
    if total_count == 0:
        return 0  # avoid division by zero
    return total_sum / total_count

nested_list = [[10, 20], [30, 40, 50], [60]]

result = average_nested(nested_list)

print("Nested list:", nested_list)
print("Average of all elements:", result)
```

```python
import numpy as np
import time
import random

# ---------- 1. Element-wise Addition ----------
size = 1000000

# Python Lists
list1 = [random.random() for _ in range(size)]
list2 = [random.random() for _ in range(size)]

start = time.time()
result_list_add = [list1[i] + list2[i] for i in range(size)]
end = time.time()

print("Python List Addition Time:", end - start)


# NumPy Arrays
arr1 = np.array(list1)
arr2 = np.array(list2)

start = time.time()
result_np_add = arr1 + arr2
end = time.time()

print("NumPy Array Addition Time:", end - start)


# ---------- 2. Element-wise Multiplication ----------

# Python Lists
start = time.time()
result_list_mul = [list1[i] * list2[i] for i in range(size)]
end = time.time()

print("Python List Multiplication Time:", end - start)


# NumPy Arrays
start = time.time()
result_np_mul = arr1 * arr2
end = time.time()

print("NumPy Array Multiplication Time:", end - start)


# ---------- 3. Dot Product ----------

# Python Lists
start = time.time()
dot_list = sum(list1[i] * list2[i] for i in range(size))
end = time.time()

print("Python List Dot Product Time:", end - start)


# NumPy Arrays
start = time.time()
dot_np = np.dot(arr1, arr2)
end = time.time()

print("NumPy Dot Product Time:", end - start)


# ---------- 4. Matrix Multiplication ----------

n = 1000

# Python Lists (matrix)
A = [[random.random() for _ in range(n)] for _ in range(n)]
B = [[random.random() for _ in range(n)] for _ in range(n)]

start = time.time()

result = [[0]*n for _ in range(n)]
for i in range(n):
    for j in range(n):
        for k in range(n):
            result[i][j] += A[i][k] * B[k][j]

end = time.time()

print("Python List Matrix Multiplication Time:", end - start)


# NumPy Matrix Multiplication
A_np = np.random.random((n,n))
B_np = np.random.random((n,n))

start = time.time()
result_np = np.dot(A_np, B_np)
end = time.time()

print("NumPy Matrix Multiplication Time:", end - start)
```
